```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

#our task is to convert into list with genres name

import ast #it helps to convert string into list
```

```python
movies.head()
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
def convert1(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords']=movies['keywords'].apply(convert1)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L      
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#it removes spaces between words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
newdf=movies[['id','title','tags']]
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
#convert list to string
```

# we are removing the stopwords and then applying VECTORIZATION

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(newdf['tags']).toarray()
```

```python
vector.shape
```

```python
vector[0]
```

```python
len(cv.get_feature_names())
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
#removing the repitetion words
```

```python
ps=PorterStemmer()
```

```python
cv.get_feature_names()
```

```python
def conv(text):
    I=[]
    for i in text.split():
        I.append(ps.stem(i))
    return " ".join(I)
```

```python
newdf['tags']=newdf['tags'].apply(conv)
```

```python
newdf.head()
```

```python
vectors=cv.fit_transform(newdf['tags']).toarray()
```

```python
newdf.head()
```

```python
cv.get_feature_names()
```

# using cosine similarities we calculate distance b/w movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
[list(similarity[0])]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=newdf[newdf['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
